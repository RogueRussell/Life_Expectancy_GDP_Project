```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('all_data.csv')
```

```python
print(df.head())
print(df.describe())
print(df.Country.value_counts())
```

```python
#Has life expectancy increased over time in the six nations?
#to show data over time, we would plot this with a lineplot
sns.lineplot(data = df, x="Year", y="Life expectancy at birth (years)", hue = 'Country')
plt.show()
```

It looks like Zimbabwe has the lowest life expectancy but a dramatic increase in the last couple years.The other 5 countries seem to have a more steady increase.

```python
#Has GDP increased over time in the six nations?
#Lineplot same as before but now looking at GDP
sns.lineplot(data = df, x="Year", y="GDP", hue = 'Country')
plt.show()
```

Seems Like US has the greatest GDP (as of 2014), with China having a large increase in GDP in the last couple years. Germany and Mexico have a slight decrease while Chile and Zimbabe are pretty consistant. 

```python
#is there a correlation between GDP and life expectancy?
#Scatterplots show correlations best
sns.scatterplot(data = df, x='GDP', y='Life expectancy at birth (years)', hue='Country')
plt.show()
```

```python
#What is the average life expectancy in these nations?
sns.barplot(data = df, x='Country', y='Life expectancy at birth (years)', hue='Country')
plt.show()
```
